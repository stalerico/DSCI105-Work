**Exercise 3**: You are given the following DataFrame representing a company's financial data:

```python
data = {
    'Year': [2018, 2019, 2020, 2021],
    'Revenue': [None, 1000000, 1500000, None],
    'Profit': [50000, None, 70000, None]
}

df = pd.DataFrame(data)
```

**Tasks**:

1. Set `Year` as the index using `set_index()`.
2. Use `reindex()` to add data for the year 2022, and fill missing values with `NaN`.
3. Confirm that the `Revenue` and `Profit` columns are `float64`.
4. Fill the missing values in `Revenue` and `Profit` with forward fill (`ffill`). Does `ffill` change the first `NaN` in a column?
5. Restore the default index and drop the old one.

````python
data = {
    'Year': [2018, 2019, 2020, 2021],
    'Revenue': [None, 1000000, 1500000, None],
    'Profit': [50000, None, 70000, None]
}

df = pd.DataFrame(data)
df = df.set_index("Year")
df = df.reindex([2018, 2019, 2020, 2021, 2022])
print(df)
print(df.info())
df = df.ffill()
print(df) # it hasnt changed the first NaN (assuming because there was no value before it)
df = df.reset_index(drop=True)
print(df)
````

**Exercise 5**: You are given the following DataFrame representing students' scores in three subjects:

```python
import pandas as pd

data = {
    'Student': ['Alice', 'Bob', 'Charlie', 'David'],
    'Math_Score': [85, 78, 92, 67],
    'English_Score': [88, 74, 90, 70],
    'Science_Score': [91, 82, 85, 65]
}

df = pd.DataFrame(data)
print(df)
```

**Tasks**:

1. Create a custom function called `calculate_total_score` that takes a row as input and returns the total score (sum of `Math_Score`, `English_Score`, and `Science_Score`).
2. Use the `apply()` function along the rows (with `axis=1`) to apply this custom function to each row and calculate the total score for each student.
3. Add the total score as a new column named `Total_Score` in the DataFrame.

````python
data = {
    'Student': ['Alice', 'Bob', 'Charlie', 'David'],
    'Math_Score': [85, 78, 92, 67],
    'English_Score': [88, 74, 90, 70],
    'Science_Score': [91, 82, 85, 65]
}

df = pd.DataFrame(data)
print(df)
def calculate_total_score(row):
    return row['Math_Score'] + row['English_Score'] + row['Science_Score']

df['Total_Score'] = df.apply(calculate_total_score, axis= 1)
print(df)
````

**Exercise**: Apply a minimum of `five` distinct pandas operations we've covered today to conduct exploratory analysis on the Titanic dataset. Your investigation should demonstrate proficiency with various DataFrame manipulation techniques.

````python
import seaborn as sns
import pandas as pd

data = sns.load_dataset('titanic')
display(data.head())

#Info
print(data.info())

# Summary statistics
print(data.describe(include='all'))

#Mean values by survival
print(data.groupby('survived').mean(numeric_only=True))

#Missing values per column=
print(data.isnull().sum())

# Passenger class distribution
print(data['class'].value_counts())

# Survival by sex
print(pd.crosstab(data['sex'], data['survived']))
````
